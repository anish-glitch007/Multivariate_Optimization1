```python
import numpy as np
import matplotlib.pyplot as plt
# import "C:/Users/Anish Mulay/Documents/2nd Year/ED5340 - DS/LAB/Lab7_ED21B011-Optimisation.ipynb"

class MultivariateOptimization:
    def __init__(self):
        self.search_dir = np.array([3,7])
        self.starting_point = np.array([2,5])
        self.new_point = self.starting_point
        alpha = 0.01
    
    def J(w1,w2):
        return (w1-10)**2 + (w2-10)**2

    def grad_J(w1,w2):
        return np.array([2*(w1-10),2*(w2-10)])
    
    def Brackets(a,b,divisions,J):
        delta = (b-a)/divisions

        w1 = a
        w2 = w1 + delta
        w3 = w2 + delta
        n = 1
        while n<=divisions:
            if w3 <= b:
                if J(w1) > J(w2) < J(w3):
                    min_range = (w1,w3)
                    # print(f"min lies between {w1} and {w3}")
                
                else:
                    w1 = w2
                    w2 = w3
                    w3 = w2 + delta
            
            else:
                print(f"no min lies between {a} and {b}")
            n+=1
        
        return (min_range)
    
    #Using interval halving method
    def IntervalHalving(a,b,e,J):
        wm = (a + b) / 2
        L = b-a
        w1 = a + L/4
        w2 = b - L/4
        while abs(L) < e:
            if J(w1) < J(wm):
                b = wm 
                wm = w1
            elif J(w2) < J(wm):
                a = wm 
                wm = w2
            else:
                a = w1
                b = w2
            L = b-a
        return wm
    


    def line_search(self,number_iterations):
        k=0
        alpha = 0.01
        self.search_dir = np.array([3,7])
        self.starting_point = np.array([2,5])
        self.new_point = self.starting_point

        for k in range(number_iterations):
                dw = alpha*self.search_dir
                self.new_point = np.add(self.starting_point,alpha * self.search_dir)

                # print(self.new_point)
                # print(np.dot(np.array(2*(self.new_point[0]-10),2*(self.new_point[1]-10) ),self.search_dir))

                if np.dot(np.array(2*(self.new_point[0]-10),2*(self.new_point[1]-10) ),self.search_dir).all() == 0:
                    # print("in condition 1")
                    return (self.new_point)
                elif J(self.new_point[0]-dw[0], self.new_point[1]-dw[1]) >= J(self.new_point[0], self.new_point[1]) <= J(self.new_point[0]+dw[0], self.new_point[1]+dw[1]):
                    # print("in condition 2")
                    return (self.new_point)

                
                elif k > number_iterations:
                    # print("in condition 3")
                    return (self.new_point)
                
                else:
                    self.starting_point = self.new_point
                k+=1
    
    
    def sgd_fixed_alpha(self,M,e1,e2):
        self.starting_point = np.array([2,5])
        self.new_point = self.starting_point
        point_list_const_alpha = np.array([])
        k=0
        alpha =0.01
        while k < M:
            self.new_point = self.starting_point - alpha*MultivariateOptimization.grad_J(*self.starting_point)
            np.append(point_list_const_alpha,self.new_point)
            k+=1
            if np.linalg.norm(MultivariateOptimization.grad_J(*self.new_point)) < e1:
                return (self.new_point)
            else:
                self.starting_point = self.new_point
        # MultivariateOptimization.plot_points(point_list= point_list_const_alpha)
    
    def sgd_variable_alpha(self,e1):
        
        alpha = np.arange(0,5,0.0001)
        self.starting_point = np.array([2,5])
        self.new_point = self.starting_point
        point_list_variable_alpha = [self.starting_point]
        x = []
        y = []
        J_list = []
        k=0
        
        for i in (alpha):
            
            self.new_point = self.starting_point - i*MultivariateOptimization.grad_J(*self.starting_point)

            point_list_variable_alpha.append(self.new_point)
            x.append(self.new_point[0])
            y.append(self.new_point[1])
            J_list.append(MultivariateOptimization.J(*self.new_point))
            # print(point_list_variable_alpha)
            # print(f"\n\npoint_list = {point_list_variable_alpha}\n\n")
            
            # MultivariateOptimization.plot_points(x_cord = x,y_cord=y,point_list = point_list_variable_alpha)
            

            if np.linalg.norm(MultivariateOptimization.grad_J(*self.new_point)) < e1:
                print("returned 1")
                return (self.new_point)
            elif np.linalg.norm(np.dot(MultivariateOptimization.grad_J(*self.new_point),MultivariateOptimization.grad_J(*self.starting_point))) < e1:
                print("returned 2")
                return (self.new_point)
            
            else:
                self.starting_point = self.new_point
        
            
            
    
    def plot_points(point_list,x_cord,y_cord):
        
        w = point_list
        x = np.arange(0,20,0.01)
        y = np.arange(0,20,0.01)
        X_cord,Y_cord = np.meshgrid(x,y)
        Z_cord = MultivariateOptimization.J(X_cord,Y_cord)
        fig,ax = plt.subplots(figsize=(6,6))
        
        ax.set(xlabel="x-axis", ylabel="y-axis")
        print(f"\n\n{point_list}\n\n")
        for point in point_list:
            # ax.plot(point,MultivariateOptimization.J(point[0],point[1]))
            ax.scatter(*point, marker="+",color = "black",s=100)

        plt.contour(X_cord,Y_cord,Z_cord)
        plt.show()
        print(w)


min_point = MultivariateOptimization()
a = min_point.sgd_variable_alpha(0.001)
b = min_point.sgd_fixed_alpha(10000,0.001,0.001)
print("Minimum point found by variable alpha methos: ",a)
print("Minimum point found by constant alpha method: ",b)

```
